```python
##Deployment
'''option_settings:
  "aws:elasticbeanstalk:container:python":
    WSGIPath: app:app'''




"""

YAML language= This file(python.config) is written in yaml.
- YAML is indentation based
- in YAML spaces matter
- Tabs are not allowed(here tabs are not allowed, only use spaces, especially 2 spaces)
- : separates key:value


1. option_settings:
- This is top level key that is used by aws elastic beanstalk
- this tells EB: I am about to define environment config options
- This is the section where you configure how your Elastic Beanstalk
 environment behaves.

2."aws:elasticbeanstalk:container:python":
- This is called namespace
- aws= provided by AWS
- elasticbeanstalk= belongs to elastic beanstalk
- container= configuration for app runtime container
- python= python specific container
-- this basically  means:“The following settings apply to the Python application
 container running on Elastic Beanstalk.”

3. Indentation is very important here:
- option_settings:
  "aws:elasticbeanstalk:container:python":

- Two spaces = child of option_settings
- This nesting tells EB:
- These settings belong under option_settings
- Wrong indentation = EB ignores it ❌

4. WSGIPath: app:app
- Web server gateway interface is the standard way that web apps talks to 
 server like:
  a. Gunicorn
  --- Gunicorn is a production-ready WSGI web server that AWS Elastic Beanstalk 
   commonly uses to serve Python web applications (like Flask) in production.
  b. uWSGI
  c. mod_wsgi
- Flask itself does NOT serve production traffic. Gunicorn + WSGI does
- Why Elastic Beanstalk needs WSGIPath? 
- Elastic Beanstalk runs: gunicorn <WSGIPath>
- So EB must know:
-- Which file contains your app
-- Which variable is the Flask app

That’s exactly what WSGIPath provides.


5.app:app
- module_name : application_object


"""
```
